```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
best_21 = pd.read_csv('your-top-songs-2021.csv')
best_20 = pd.read_csv('your-top-songs-2020.csv')
best_19 = pd.read_csv('your-top-songs-2019.csv')
best_18 = pd.read_csv('your-top-songs-2018.csv')
best_17 = pd.read_csv('your-top-songs-2017.csv')
```

```python
top_tracks = [best_21, best_20, best_19, best_18, best_17]

top_df = pd.concat(top_tracks)
```

```python
# Find the 10 Least popular songs on spotify of my top songs
sorted_df = top_df.sort_values('Popularity', ascending=True).head(10)  
sorted_df
```

```python
# Find 10 most popular songs with popularity > 75 of my top songs

most_popular = top_df.query('Popularity>75', inplace=False).sort_values('Popularity',ascending=False)
most_popular.head(10)
```

```python
# Let's Set "release_date" as index of dataset

top_df.set_index('Album Release Date',inplace=True)
top_df.index = pd.to_datetime(top_df.index)
top_df.head()
```

```python
# Convert the "duration_ms" to "duration" in seconds

top_df['Duration'] = top_df['Track Duration (ms)'].apply(lambda x: round(x/1000))
top_df.drop('Track Duration (ms)', inplace =True, axis = 1)
top_df.head()
```

##  Correlation of the Columns 

```python
top_df_temp = top_df.drop(['Key','Mode','Explicit'], axis = 1) # Dropping unwanted columns from the dateset
corr_df = top_df_temp.corr(method='pearson')

# Plotting the correlation values 
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True, fmt = "0.1g",vmin=-1,vmax=1,center=0, cmap="inferno",linecolor='black',linewidths=1)
heatmap.set_title("Correaltion HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation =90)

plt.show()
%matplotlib inline
```

```python
# Regression Plot using 40 percent of the data

sample_df = top_df.sample(int(0.4*len(top_df))) # Sampling data
print(len(sample_df)) # Length of sampled data
```

```python
# Loudness Vs Energy Correlation

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="Energy", y="Loudness", color="c").set_title("Loudness Vs Energy Correlation")
```

## Popularity Vs Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="Acousticness", y="Popularity", color="b").set_title("Popularity Vs Acousticness Correlation")
```

## Danceability vs Instrumentalness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="Instrumentalness", y="Danceability", color="orange").set_title("Instrumentalness Vs Danceability Correlation")
```

## Tempo Vs Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, x="Tempo", y="Energy", color="purple").set_title("Tempo Vs Energy Correlation")
```

## Average Length of Songs Based on Year

```python
top_df['Dates'] = top_df.index
years = top_df.Dates.dt.year

# Distribution plot for number of songs per year
sns.displot(years, discrete=True, aspect=2, height=5, kind="hist").set(title="Number of songs per year")
```

```python
total_dr = top_df.Duration
fig, ax = plt.subplots(figsize=(18,7))
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth=False).set(title = "Years Vs Duration")
plt.xticks(rotation= 90)

plt.show()
%matplotlib inline
```

```python
# Bar Plot of duration of songs in different genres

plt.figure(figsize=(12,8))
sns.barplot(y ='Artist Genres', x='Duration', data = top_df).set(title = "Duration of the Songs in Different Genres")
plt.ylabel("Gneres")
plt.xlabel("Duration in seconds")

plt.show()
%matplotlib inline
```
